```python
%%capture cap --no-stderr
# Path to the trained model and the image
model_path = os.path.join(os.getenv("HOME"), "/content/bestv8.pt")
image_path = "/content/djj.jpg"  # Change this to the path of your custom photo

# Command to perform detection
!yolo task=detect mode=predict model="{model_path}" source="{image_path}" save=True

# Assuming the output is saved in a directory named predict7
output_dir = "/content/runs/detect/predict3"

# Write output to a text file
output_text_path = os.path.join(output_dir, "output.txt")
with open(output_text_path, 'w') as f:
    f.write(cap.stdout)

# Display the output image
output_image_path = os.path.join(output_dir, os.path.basename(image_path))
display(Image(filename=output_image_path))

print(f"Detection results saved to {output_text_path}")
```

```python
def extract_image_info(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith('image 1/1'):
                return line
    return None
def parse_buildings(image_info):
    # This regex extracts the number of buildings from the line
    import re
    match = re.search(r"(\d+) Buildings", image_info)
    if match:
        return int(match.group(1))
    return 0

def categorize_environment(buildings_count):
    urban_threshold = 50  # More than 50 buildings means urban
    if buildings_count > urban_threshold:
        return "Urban"
    return "Rural"
# Path to the output file
output_file_path = "/content/runs/detect/predict3/output.txt"

# Extract the relevant image info
image_info = extract_image_info(output_file_path)
if image_info:
    buildings_count = parse_buildings(image_info)
    category = categorize_environment(buildings_count)
    print(f"Environment is categorized as: {category}")
else:
    print("No image info found.")
```
